# Relacion entre la edad y el sexo de la victima

```python
#Importo bibliotecas necesarias.
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Base de datos.
url_original='https://raw.githubusercontent.com/ianCristianAriel/proc_datos_proyecto_final_G3/visualizacion/datos/procesados/limpieza01_argentina_delitos.csv'
url_10='https://raw.githubusercontent.com/ianCristianAriel/proc_datos_proyecto_final_G3/visualizacion/datos/procesados/preparacion10_datos_edad_sexo_victima.csv'
```

```python
#Leo y reemplazo los nombres
df_original= pd.read_csv(url_original)
df_sexo_edad_victimas= pd.read_csv(url_10)
```

```python
#Aca reemplazo los valores sexo_victima
df_sexo_edad_victimas['sexo_victima'] = df_original['sexo_victima']
#Aca reemplazo los valores tr_edad_victima
df_sexo_edad_victimas['edad_victima'] = df_original['tr_edad_victima']
#Aca Agrego la columna tipo_persona
df_sexo_edad_victimas['tipo_persona'] = df_original['tipo_persona']
#Filtro por victima
df_victima = df_sexo_edad_victimas[df_sexo_edad_victimas['tipo_persona'] == 'Víctima']
# Convertir la columna 'edad_victima' a valores enteros
df_sexo_edad_victimas['edad_victima'].replace(to_replace='string', value='', inplace=True)
# Filtrar por valores válidos de edad (valores numéricos)
df_victima_edad_int = df_victima[df_victima['edad_victima'].notnull()]
```

```python

# Agrupa los datos por edad y sexo de la víctima y cuenta la cantidad de registros en cada grupo
grupo_edad_sexo = df_victima_edad_int.groupby(['edad_victima', 'sexo_victima']).size().unstack()

# Genera el gráfico de barras horizontales
grupo_edad_sexo.plot(kind='barh', stacked=True)

# Configura los ejes y el título del gráfico
plt.xlabel('Cantidad de víctimas')
plt.ylabel('Edad de la víctima')
plt.title('Relación entre la edad y el sexo de la víctima')

# Muestra el gráfico
plt.show()
```

# Relacion entre la edad y el sexo de la inculpado

```python
#Base de datos
url_11='https://raw.githubusercontent.com/ianCristianAriel/proc_datos_proyecto_final_G3/visualizacion/datos/procesados/preparacion11_edad_inculpado.csv'
```

```python
#Leo y reemplazo los nombres
df_sexo_edad_inculpados= pd.read_csv(url_11)
```

```python
# Reemplazar los valores de sexo_inculpado
df_sexo_edad_inculpados['sexo_inculpado'] = df_original['sexo_inculpado']
# Reemplazar los valores de tr_edad_inculpado
df_sexo_edad_inculpados['edad_inculpado'] = df_original['tr_edad_inculpado']
# Agregar la columna tipo_persona
df_sexo_edad_inculpados['tipo_persona'] = df_original['tipo_persona']
# Filtrar por inculpado
df_inculpado = df_sexo_edad_inculpados[df_sexo_edad_inculpados['tipo_persona'] == 'Imputado']
# Convertir la columna 'edad_inculpado' a valores enteros
df_sexo_edad_inculpados['edad_inculpado'].replace(to_replace='string', value='', inplace=True)
# Filtrar por valores válidos de edad (valores numéricos)
df_inculpado_edad_int = df_inculpado[df_inculpado['edad_inculpado'].notnull()]
```

```python
# Agrupa los datos por edad y sexo del inculpado y cuenta la cantidad de registros en cada grupo
grupo_edad_sexo = df_inculpado_edad_int.groupby(['edad_inculpado', 'sexo_inculpado']).size().unstack()

# Genera el gráfico de barras horizontales
grupo_edad_sexo.plot(kind='barh', stacked=True)

# Configura los ejes y el título del gráfico
plt.xlabel('Cantidad de inculpados')
plt.ylabel('Edad del inculpado')
plt.title('Relación entre la edad y el sexo del inculpado')

# Muestra el gráfico
plt.show()
```
